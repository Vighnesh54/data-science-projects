#### Task 2: Dimensionality Reduction with PCA:

##### 1.Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)
```

```python
pca= PCA()
pca_data = pca.fit_transform(scaled_data)
```

```python
pca_df =pd.DataFrame(
    pca_data,
    columns=[f'PC{i+1}' for i in range(pca_data.shape[1])]
)
```

```python
explained_variance= pca.explained_variance_ratio_
```

```python
print("Explained variance ratio of each Principal Component:")
print(explained_variance)
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(1, len(explained_variance)+1), explained_variance, marker='o', linestyle='--')
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.title("PCA - Explained Variance Ratio")
plt.show()
```

##### 2. Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.

```python
cumulative_variance= np.cumsum(explained_variance)
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o', linestyle='--')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Cumulative Explained Variance by PCA Components")
plt.axhline(y=0.90, color='r', linestyle='--', label="90% threshold")
plt.legend()
plt.show()
```

##### Scree Plot
Plots the explained variance ratio of each PC.
Look for the “elbow point” where adding more components gives only a small increase in explained variance.

##### Cumulative Explained Variance Plot
Shows how much total variance is explained as we add more components.
Typically, we select the number of PCs that explain at least 90–95% of the variance.

### Task 3: Clustering with Original Data:

##### 1. Apply a clustering algorithm (e.g., K-means) to the original dataset.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
x = df.drop(columns=['Type'])
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(x_scaled)
```

```python
df['Cluster'] = clusters
df.head()
```

```python
print(df['Cluster'].value_counts())
```

```python
from sklearn.metrics import confusion_matrix,classification_report
from scipy.optimize import linear_sum_assignment
```

```python
true_labels = df['Type']
```

```python
cm = confusion_matrix(true_labels,clusters)
```

```python
row_ind, col_ind = linear_sum_assignment(-cm)
mapping = dict(zip(col_ind, row_ind + 1))
```

```python
mapped_clusters = np.array([mapping[c] for c in clusters])
```

```python
print("Confusion Matrix after mapping:\n", confusion_matrix(true_labels, mapped_clusters))
print("\nClassification Report:\n", classification_report(true_labels, mapped_clusters,zero_division=0))
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=x_scaled[:,0], y=x_scaled[:,1], hue=clusters, palette="viridis")
plt.title("K-means Clustering on Wine Dataset")
plt.xlabel("Feature 1 (Alcohol)")
plt.ylabel("Feature 2 (Malic)")
plt.show()
```

##### 2. Visualize the clustering results using appropriate plots.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
```

```python
# Plot clusters
plt.figure(figsize=(8,6))
sns.scatterplot(x=x_pca[:,0], y=x_pca[:,1], hue=mapped_clusters, palette="Set1", s=70)
plt.title("Clustering Results Visualized with PCA (2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Cluster")
plt.show()
```

##### 3. Evaluate the clustering performance using metrics such as silhouette score or Davies–Bouldin index.

```python
from sklearn.metrics import silhouette_score,davies_bouldin_score
```

```python
silhouette_avg = silhouette_score(x_scaled,clusters)
db_index = davies_bouldin_score(x_scaled,clusters)
```

```python
print("Silhouette Score:", silhouette_avg)
print("Davies-Bouldin Index:", db_index)
```

The Silhouette Score indicates that the clusters are well separated, while the Davies–Bouldin Index confirms the compactness and separation of the clusters.

##### 4. Analyze the trade-offs between using PCA and clustering directly on the original dataset.

##### Advantages of Using PCA before Clustering:
Dimensionality Reduction: PCA reduces the number of features, lowering computational cost and making clustering faster, especially on large datasets.

Noise Filtering: PCA removes less informative variance, helping focus on the most meaningful patterns.

Visualization: By reducing data to 2 or 3 dimensions, PCA makes it possible to visualize clusters clearly.

##### Disadvantages of Using PCA:

Information Loss: Some variance is discarded during PCA, which can slightly reduce clustering accuracy (as seen in the Davies-Bouldin Index being a bit higher after PCA).

Interpretability: PCA components are linear combinations of features, so the original meaning of features is lost, making results harder to interpret.

##### Clustering on Original Dataset:

Preserves all information and original feature meanings.

Can sometimes give slightly better cluster separation (as in this case, original DBI = 0.5496 vs PCA = 0.5977).

But, high dimensionality may introduce noise, redundancy, and higher computational cost

##### Conclusion:
Using PCA involves a trade-off: it simplifies clustering and improves visualization while maintaining similar quality, but at the cost of minor information loss and reduced interpretability. The original dataset may give slightly better separation, but PCA is highly beneficial when dealing with large, complex datasets.

##### 2. Discuss the practical implications of using PCA and clustering in data analysis.

1.Dimensionality Reduction:
PCA reduces high-dimensional data into fewer principal components, which simplifies the dataset while retaining most of the important information. This helps reduce computational cost and makes clustering faster and more efficient.

2.Noise Reduction:
By focusing only on the most significant components, PCA removes irrelevant or redundant features. This improves clustering accuracy because the algorithm focuses on meaningful patterns rather than noise.

3.Visualization of Clusters:
PCA transforms complex datasets into 2D or 3D space, making it possible to visually inspect clusters. This is especially useful when presenting results to non-technical stakeholders.

4.Improved Efficiency in Clustering:
High-dimensional datasets often make clustering algorithms (like K-means) perform poorly due to the "curse of dimensionality." PCA addresses this by reducing dimensions, leading to more stable and compact clusters

5.Trade-off Between Accuracy and Interpretability:
While PCA simplifies data and improves clustering efficiency, some original information is lost. This means the clusters may be easier to visualize but slightly less accurate compared to clustering on the original dataset.

6.Applicability in Real-World Scenarios:
In practical applications such as market segmentation, image recognition, bioinformatics, or text analysis, PCA combined with clustering provides a balance between accuracy, speed, and interpretability.
